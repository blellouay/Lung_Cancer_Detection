```python
from src.utils.helpers import read_json, show_image_plt
```

```python
annotations=read_json(r"Data2\tunisia-camera-shot-dataset\annotations.json")
```

```python
import cv2
image=cv2.imread("data2/"+annotations[0]["relative_path"])

print(image.shape)
show_image_plt(image)
```

```python
base_path="Data2"
for idx,annot in enumerate(annotations):
    image_correct_path=f"{base_path}/{annot['relative_path']}"
    image=cv2.imread(image_correct_path)
    print(image_correct_path , annot['class'])
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    cv2.putText(image, f"class: {annot['class']}",(20, 40), cv2.FONT_HERSHEY_SIMPLEX,  2.0,(255, 0,0), 2,cv2.LINE_AA )
    show_image_plt(image)
    if idx>6: break
```
